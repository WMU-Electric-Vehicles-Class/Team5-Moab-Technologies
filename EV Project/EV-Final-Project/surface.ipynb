```python
import numpy as np
import pandas as pd
from scipy.interpolate import LinearNDInterpolator
from matplotlib import pyplot as plt
```

Front motor data

```python
front = pd.read_excel("data/Audi_e_Tron_Motors_ResultsForMaps.xlsx", "APA250-Maps")
front
```

```python
x1 = front["SPEED [rad/s]"]
y1 = front["TORQUE [Nm]"]
z1 = front["EFFY [%]"]
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.view_init(elev=90, azim=-90)

surf = ax.plot_trisurf(x1, y1, z1, cmap="coolwarm", linewidth=0)
```

Rear motor data

```python
rear = pd.read_excel("data/Audi_e_Tron_Motors_ResultsForMaps.xlsx", "AKA320-Maps")
rear
```

```python
x2 = rear["SPEED [rad/s]"]
y2 = rear["TORQUE [Nm]"]
z2 = rear["EFFY [%]"]
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.view_init(elev=90, azim=-90)

surf = ax.plot_trisurf(x2, y2, z2, cmap="coolwarm", linewidth=0)
```

Front interpolation

```python
interp_front = LinearNDInterpolator((x1, y1), z1)

xnew = np.linspace(min(x1), max(x1), 500)
ynew = np.linspace(min(y1), max(y1), 500)
xnew, ynew = np.meshgrid(xnew, ynew)
znew = interp_front(xnew, ynew)

fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.view_init(elev=90, azim=-90)
surf = ax.plot_surface(xnew, ynew, znew, cmap="coolwarm",
                       linewidth=0, antialiased=False)
```

Interpolate single point

```python
x_test = 1000
y_test = 100
z_test = interp_front(x_test, y_test)
float(z_test)
```

Rear interpolation

```python
interp_rear = LinearNDInterpolator((x2, y2), z2)

xnew = np.linspace(min(x2), max(x2), 500)
ynew = np.linspace(min(y2), max(y2), 500)
xnew, ynew = np.meshgrid(xnew, ynew)
znew = interp_rear(xnew, ynew)

fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.view_init(elev=90, azim=-90)
surf = ax.plot_surface(xnew, ynew, znew, cmap="coolwarm",
                       linewidth=0, antialiased=False)
```

Interpolate single point

```python
x_test = 1000
y_test = 100
z_test = interp_rear(x_test, y_test)
float(z_test)
```
