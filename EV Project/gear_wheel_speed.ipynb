```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# kph
#Wo1/Win1 = 15
#Wo2/Win2 = 8

# x data in kph
Wo = np.linspace(0, 97, 10) #kph
Woo = np.linspace(97, 230, 10) #kph

# y data in rad/h
Wo1 = np.linspace(0, 230, 10)
Win1 = Wo/15
Win2 = Woo/8
```

```python
plt.plot(Wo, Win1, linestyle = ':', label = '1st gear')
plt.plot(Woo, Win2, linestyle = '-.', label = '2nd gear')
plt.title('2-speed Transmission: Angular Gear Speed')
plt.xlabel('Speed [kph]')
plt.ylabel('Electric Motor: Angular Speed [rad/h]')
plt.legend()
plt.grid()
plt.show()
```
